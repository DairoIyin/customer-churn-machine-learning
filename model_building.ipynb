```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import NearMiss
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv("tel_churn.csv")
df.head()
```

```python
df.columns.values
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
x=df.drop('Churn',axis=1)
```

```python
x
```

```python
y=df['Churn']
y
```

### Train Test Split

 Split the training data. 70% of our telecommunications dataset is for training, and 30% for testing.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

## Decision Tree Classifier

```python
#Define the model
#gini is the selected impurity measure
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

fit the training data

```python
model_dt.fit(x_train,y_train)
```

make a prediction

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

As you can see that the accuracy is quite low, and as it's an imbalanced dataset, we shouldn't consider Accuracy as our metrics to measure the model in imbalanced datasets.
Hence, we need to check recall, precision & f1 score for the minority class, and it's quite evident that the precision, recall & f1 score is too low for Class 1, i.e. churned customers.

Hence, moving ahead to perform Undersampling

## Balancing the Data

The undersampling method chosen is NearMiss version 3


```python
#Create a NearMiss3 object
nm = NearMiss(version=3, n_neighbors=3) 
```

```python
xr_train,xr_test,yr_train,yr_test=train_test_split(x, y,test_size=0.3)
```

```python
#Resample the dataset
X_resampled, y_resampled = nm.fit_resample(xr_train,yr_train)
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(criterion='gini',random_state = 42),
        'params' : {
            'n_estimators': [1,5,10],
            'max_depth':[6,11,16],
            'min_samples_leaf':[8,10,15]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear',random_state = 42),
        'params': {
            'C': [1,5,10]
        }
    },
    'decision_tree': {
        'model': DecisionTreeClassifier(criterion = "gini",random_state = 42),
        'params': {
            'max_depth':[6,11,16],
            'min_samples_leaf':[8,10,15]
        }
    },
    'naive_bayes': {
        'model': GaussianNB(),
        'params': {
            'random_state':[42,100]
        }
    }

}
```

```python
scores = []

for model_name, mp in model_params.items():
    clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(X_resampled, y_resampled)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_,
        
    })
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params',])
df
```

```python
model_dt_nm=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
clf= GridSearchCV(model_dt_nm,)
```

```python
model_dt_nm.fit(X_resampled,y_resampled)
yr_predict = model_dt_nm.predict(xr_test)
model_score_r = model_dt_nm.score(xr_test, yr_test)
print(model_score_r)
print(metrics.classification_report(yr_test, yr_predict))
```

```python
print(metrics.confusion_matrix(yr_test, yr_predict))
```
